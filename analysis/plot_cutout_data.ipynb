# A simple script to plot the science images used for the modeling

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

names = ['J0407-5006', 'J0602-4335', 'J0806+2006', 'J1001+5027', 'J1442+4055',
         'J1515+1511', 'J1620+1203', 'J2325-5229']

filter = 'F160W'

fig, ax = plt.subplots(int(len(names) / 2), 2, figsize = (10, 20))

for i, name in enumerate(names):
    filename = f'../cutout_data/{name}/{filter}/{name}_{filter}_cutout.fits'

    with fits.open(filename) as hdul:
        data = hdul[0].data

    ax = ax.flatten()

    ax[i].set_title(f'{name}')
    ax[i].imshow(np.log10(data), origin='lower', cmap='cubehelix')
```
